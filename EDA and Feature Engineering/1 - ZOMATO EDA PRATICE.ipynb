```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# merging to add country name to the DF
data = pd.merge(df,df_country,on='Country Code',how='left')
data.head()
```

```python
data.dtypes
```

```python
#getting the country names from the country column
country_name= data.Country.value_counts().index
country_name
```

```python
#getting the counts of country names from country column
country_value= data.Country.value_counts().values
country_value
```

```python
#using pie chart to visualize the top 3 country
plt.pie(country_value[:3], labels=country_name[:3], autopct= '%1.2f%%')
```

#### Observation:
 most of the transactions of zomato are from India followed by United States and United Kingdom

```python
rating = data.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating counts'})
rating
```

#### Observation
1. when rating is between 4.5 - 4.9 ---> Excellent
2. when rating is between 4.0 - 4.4 ---> VeryGood
3. when rating is between 3.5 - 3.9 ---> Good
4. when rating is between 2.5 - 3.4 ---> Average
5. when rating is between 2.4 - 1.8 ---> poor


```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,6)
sns.barplot(x='Aggregate rating', y='Rating counts',data=rating,hue= 'Rating color',palette=['#BBc3CB','red','orange','yellow','green','green'])
```

#### Observation
1. not rated is very high

2.The max rating are mostly between the yellow and orange which are Good and Average
