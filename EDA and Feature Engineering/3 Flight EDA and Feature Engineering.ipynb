```python
#importing basics libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading the train dataset
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
#loading the test dataset
test_df=pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = train_df.append(test_df)
final_df.tail()
```

```python
final_df.info()
```

```python
## feature engineering process
# seperating date , month, year from date of journey column
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] =final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

```python
# using lambda
#final_df['Date']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
```

```python
# change from object to integer
final_df[['Date','Month','Year']]= final_df[['Date','Month','Year']].astype(int)
```

```python
final_df.info()
```

```python
#dropping the date of journey column after seperating it
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
#final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
# seperating the hour and min from Arrival_time column
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df['Arrival_hour']= final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']= final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head(1)
```

```python
#changing to integer
final_df[['Arrival_hour','Arrival_min']]= final_df[['Arrival_hour','Arrival_min']].astype(int)
```

```python
final_df.info()
```

```python
#dropping the arrival_time column after seperating it
final_df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
# seperating the departure time into hour and min  removing ':'
final_df['Dep_hour']= final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min']= final_df['Dep_Time'].str.split(':').str[1]

final_df[['Dep_hour','Dep_min']]= final_df[['Dep_hour','Dep_min']].astype(int)
final_df.drop("Dep_Time",axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
#mapping the Total stops to numerical value
final_df['Total_Stops'] =final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':2})
```

```python
final_df.head()
```

```python
#dropping the Route column since we already the total_stops ,source and destination columns
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
# seperating text value 'h' and 'm' from the duration time column
final_df['duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.head(1)
```

```python
# there are two insignificant value for the new duration hour
final_df[final_df['duration_hour']== '5m']
```

```python
# dropping the insignificant value
final_df.drop([6474,2660],axis=0,inplace=True)
```

```python
#checking to confirm if dropped
final_df[final_df['duration_hour']== '5m']
```

```python
# changing type from object to integer
final_df['duration_hour'] = final_df['duration_hour'].astype('int')
```

```python
final_df['duration_hour']*60
```

```python
#convert for minute and also remove the 'm' text value
final_df['duration_min'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df.head(1)
```

```python

#final_df['duration_min']= final_df['duration_min'].astype(int)
```

```python
# unfortunately there were 1283 nan while trying to covert from object to integer
final_df['duration_min'].isnull().sum()
```

```python
final_df['duration_min'].unique()
```

```python
# setting the nan value to be 0 obviously from values having only hour 'h' and no need for minutes
final_df['duration_min'] = final_df['duration_min'].fillna(0)
final_df['duration_min'].isnull().sum()
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```
